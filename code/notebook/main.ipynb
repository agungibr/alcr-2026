# preprocessing

### cleaning

```python
df['timestamp'] = pd.to_datetime(
    df['timestamp'],
    format='%Y-%m-%dT%H:%M:%S.%f%z'
)

label['started_at'] = pd.to_datetime(label['started_at'])
label['finished_at'] = pd.to_datetime(label['finished_at'])

df['timestamp'] = df['timestamp'].dt.tz_convert('Asia/Tokyo')
label['started_at'] = label['started_at'].dt.tz_convert('Asia/Tokyo')
label['finished_at'] = label['finished_at'].dt.tz_convert('Asia/Tokyo')
```

```python
df = df.drop(columns=['name'])
label = label.drop(columns=['Unnamed: 0.1', 'Unnamed: 0'])

a = label[pd.notnull(label['started_at'])]
b = a[pd.notnull(a['finished_at'])]

labelx = b["deleted_at"].dropna()
labelx.value_counts().sum()

label_clean = b[~b["deleted_at"].isin(labelx)]

label_clean = label_clean.loc[label_clean['activity'] == 'Location']

label_clean = label_clean.drop(columns=['deleted_at', 'updated_at', 'activity'])
```

```python
df = df.sort_values('timestamp').reset_index(drop=True)
label_clean = label_clean.sort_values('started_at').reset_index(drop=True)
```

```python
# ambil rentang label
start_label = label_clean['started_at'].min()
end_label = label_clean['finished_at'].max()

# potong RSSI ke rentang label
df = df[(df['timestamp'] >= start_label) & (df['timestamp'] <= end_label)].copy()

print(df['timestamp'].min(), df['timestamp'].max())
```

```python
import numpy as np

rooms = []
lbl = label.to_dict('records')
j = 0
n = len(lbl)

for ts in df['timestamp']:
    # geser interval label sampai finished_at >= ts
    while j < n and ts > lbl[j]['finished_at']:
        j += 1

    if j < n and lbl[j]['started_at'] <= ts <= lbl[j]['finished_at']:
        rooms.append(lbl[j]['room'])
    else:
        rooms.append(np.nan)

df['room'] = rooms

# buang RSSI tanpa label
df1 = df.dropna(subset=['room']).reset_index(drop=True)
```

```python
WINDOW = '45s'

df['window_start'] = df['timestamp'].dt.floor(WINDOW)
```

```python
room_per_window = (
    df.groupby('window_start')['room']
      .nunique()
      .value_counts()
)

room_per_window
```

```python
valid_windows = (
    df.groupby('window_start')['room']
      .nunique()
      .reset_index()
)

valid_windows = valid_windows[valid_windows['room'] == 1]['window_start']

df = df[df['window_start'].isin(valid_windows)].reset_index(drop=True)
```

```python
df.groupby('window_start')['room'].nunique().value_counts()
```

```python
# RSSI matrix: window x beacon
rssi_matrix = (
    df
    .groupby(['window_start', 'mac address'])['RSSI']
    .mean()              # agregasi dasar untuk visualisasi matrix
    .unstack('mac address')
    .sort_index()
)
```

```python
print(rssi_matrix.shape)
rssi_matrix.iloc[:5, :5]
```

```python
agg_funcs = ['mean', 'std', 'var', 'min', 'max', 'median', 'sum', 'count']

feat = (
    df
    .groupby(['window_start', 'room', 'mac address'])['RSSI']
    .agg(agg_funcs)
    .unstack('mac address')
)

# flatten kolom: {mac}_{stat}
feat.columns = [f'{mac}_{stat}' for stat, mac in feat.columns]
feat = feat.reset_index()
```

```python
feat['hour'] = feat['window_start'].dt.hour
feat['minute'] = feat['window_start'].dt.minute
feat['second'] = feat['window_start'].dt.second
```

```python
feat = feat.fillna(0)
```

```python
X = feat.drop(columns=['window_start', 'room'])
y = feat['room']
```

```python
X.columns = (
    X.columns
    .str.replace(':', '_', regex=False)
    .str.replace('-', '_', regex=False)
    .str.replace(' ', '_', regex=False)
)
```

```python
min_samples = 2
valid_rooms = y.value_counts()[y.value_counts() >= min_samples].index

mask = y.isin(valid_rooms)
X = X[mask]
y = y[mask]
```

```python
print(X.shape, y.shape)
X.iloc[:2, :5]
y.value_counts()
```

```python
from sklearn.preprocessing import LabelEncoder

le = LabelEncoder()
y_enc = le.fit_transform(y)
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(
    X, y_enc,
    test_size=0.3,
    random_state=42,
    stratify=y_enc
)
```

```python
import lightgbm as lgb

model = lgb.LGBMClassifier(
    objective='multiclass',
    num_class=len(le.classes_),
    n_estimators=300,
    learning_rate=0.05,
    num_leaves=31,
    subsample=0.8,
    colsample_bytree=0.8,
    random_state=42,
    n_jobs=-1
)

model.fit(X_train, y_train)
```

```python
from sklearn.metrics import accuracy_score, classification_report

y_pred = model.predict(X_test)

print("Accuracy:", accuracy_score(y_test, y_pred))
print(classification_report(y_test, y_pred, target_names=le.classes_))
```
